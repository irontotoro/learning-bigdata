```python
import csv
with open('seoul.csv','r',encoding='cp949') as f:
    data=csv.reader(f,delimiter=',')
    header=next(data)
    max_temp=-1000
    max_date=''

    for row in data:
        if row[-1]=='':
            row[-1]=-999
        row[-1]=float(row[-1])
            
        if row[-1]>max_temp:
            max_temp=row[-1]
            max_date=row[0]
    print("최고온도:",max_temp,'그 날짜:',max_date)
```

```python
import csv
with open('seoul.csv','r',encoding='cp949') as f:
    data=csv.reader(f,delimiter=',')
    header=next(data)
    max_diff=0
    max_date=''
    high, low=0,0
    for row in data:
        if row[-1]!=''and row[-2]!='':
            diff=float(row[-1])-float(row[-2])
            if diff>max_diff:
                max_diff=diff
                max_date=row[0]
                high=row[-1]
                low=row[-2]
    print(f"최고차이: {max_diff} 날짜:{max_date} {high} {low}")
```

```python
import csv
import matplotlib.pyplot as plt
nomalized_high=[]
nomalized_low=[]
with open('seoul.csv','r') as f:
    data=csv.reader(f)
    header=next(data)
    for row in data:
        if row[-1]!='':
            if int(row[0].split('-')[0])>=1983:
                if row[0].split('-')[1]=='02'and row[0].split('-')[2]=='14':
                    nomalized_high.append(float(row[-1])-float(row[2]))
                    nomalized_low.append(float(row[-2])-float(row[2])) 
plt.title("TEMPERATURE")
plt.rc("font",family='Times New Roman')
plt.plot(nomalized_high,'r',label='normed Max temp')
plt.plot(nomalized_low,'b',label='normed Min temp')
plt.legend()
plt.show()
```
