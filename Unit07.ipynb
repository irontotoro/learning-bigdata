```python
import csv
import matplotlib.pyplot as plt
with open('age.csv') as f:
    data=csv.reader(f)
    header=next(data)
    for row in data :
        if '신도림'in row[0]:
            result=[]
            for i in row[3:]:
                result.append(int(i))
print(result)
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
with open('age.csv') as f:
    data=csv.reader(f)
    header=next(data)
    for row in data :
        if '신도림'in row[0]:
            result=list(map(int,row[3:]))
plt.title('PEOPLE')
print(result)
plt.plot(result)
plt.show()
```
