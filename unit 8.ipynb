```python
import csv
import matplotlib.pyplot as plt
f=open('age.csv')
data=csv.reader(f)
result=[]
name=input()

for row in data :
    if name in row[0]:
        result=list(map(int,row[3:]))

plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title(name+'지역의 인구구조')
plt.bar(range(101),result,color='b')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt 
import matplotlib 
from matplotlib import rc
import seaborn as sns
%matplotlib inline

rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False

f=open('age.csv',encoding='cp949')
data=csv.reader(f)
result=[]
name=input()
for row in data :
    if name in row[0]:
        result=list(map(int,row[3:]))
slen=10
sresult=[0]*slen
for i in range(slen):
    sresult[i]=sum(result[i*10:(i+1)*10])  
plt.style.use('ggplot')

plt.title(name+'지역의 인구구조')
plt.bar(range(10),sresult,color='b')
plt.show()
```
