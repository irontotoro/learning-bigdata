```python
import csv
import matplotlib.pyplot as plt
with open('gender.csv')as f:
    data=csv.reader(f)
    header=next(data)
    name=input('동을 입력해주세요: ')
    for row in data:
        if name in row[0]:
            m=row[3:104]
            f=row[106:]
            break
for i in range(len(m)):
    m[i]=m[i].replace(',','')
    f[i]=f[i].replace(',','')
m=list(map(int,m))

#for i in range(len(m)):
 #   m[i]=-m[i]
    
#def min_int(string):
    #return -int(string)
    
m=[-i for i in m]

f=list(map(int,f))
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+'지역의 남녀성별인구분포')
plt.barh(range(101),m,color='purple')
plt.barh(range(101),f,color='yellow')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
male=[]
female=[]
with open('gender.csv')as f:
    data=csv.reader(f)
    header=next(data)
    name=input('동을 입력해주세요: ')
    for row in data:
        if name in row[0]:
            for i in row[3:104]:
                male.append(-int(i.replace(',','')))
            for j in row[106:]:
                female.append(int(j.replace(',','')))
            break
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+'지역의 남녀성별인구분포')
plt.barh(range(101),male,color='springgreen')
plt.barh(range(101),female,color='violet')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
with open('subwayfee.csv')as file:
    data=csv.reader(file)
    header=next(data)
    print(header)
    max_rate=0
    max_station= ''
    for row in data:
        for i in range(4,8):
            row[i]=int(row[i])
        if row[6]!=0 and(row[6]+row[4])>1e5:
            rate=row[4]/(row[6]+row[4])
            if rate>max_rate:
                max_rate=rate
                max_station=row[1]+row[3]
            if rate>0.94:
                print(row,round(rate,2))
print(f"최대 rate 역은 {max_station}입니다")
```

```python
import csv
import matplotlib.pyplot as plt
with open('subwayfee.csv')as file:
    data=csv.reader(file)
    header=next(data)
    plt.rc('font',family='Malgun Gothic')
    result=[]
    for row in data:
        passengers=0
        for i in range(4,8):
            row[i]=int(row[i])
            passengers+=row[i]
        name=row[1]+row[3]
        result.append([passengers,name])
    result=sorted(result,reverse=True)[:10]
    label=[]
    passenger_cnt=[]
    for i in range(10):
        label.append(result[i][1])
        passenger_cnt.append(result[i][0])
    plt.pie(passenger_cnt,autopct='%2f',labels=label)
    plt.title('서울 바쁜역')
    plt.show()
    
    
```
