```python
import csv

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
A=10
offset=2
period=100
noise_A=3
n_samples=5*period
inst_r=9
times=np.arange(n_samples).astype(float)
sig= A*np.sin(2*np.pi*times/period)
sig+=offset
sig_off=sig+offset
noise=noise_A*np.random.normal(size=n_samples)
sig_noise=sig_off+noise
sig_noise2=sig_noise[:]
sig_noise2[sig_noise2>inst_r]=inst_r
sig_noise2[sig_noise2<-inst_r]=-inst_r
fig,axs=plt.subplots(4,1,figsize=(25,10))
matplotlib.style.use('ggplot')
axs[0].plot(times,sig)
axs[0].set_title('Signal')
axs[1].plot(times,sig_off)
axs[1].set_title('Signal+Offset')
axs[2].plot(times,sig_noise)
axs[2].set_title('Signal+Off+Noise')
axs[3].plot(times,sig_noise2)
axs[3].set_title('Clamped')
```
